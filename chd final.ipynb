```python
di={'30-40':0,'41-50':0,'51-65':0}
l=df1.shape[0]
i=0
while i<l:
    if ((df1.age[i]>30 and df1.age[i]<=40) and df1.TenYearCHD[i]==1):
        di['30-40']+=1
    elif((df1.age[i]>40 and df1.age[i]<=50) and df1.TenYearCHD[i]==1):
        di['41-50']+=1
    elif((df1.age[i]>51 and df1.age[i]<=65) and df1.TenYearCHD[i]==1):
        di['51-65']+=1
    i+=1

labels = di.keys()
sizes = di.values()

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%')
ax1.axis('equal')
plt.title('age grp who will have CHD in a 10 year tenure ')


plt.show()
```

```python
di={'has diabetes':0,'no diabetes':0}

for i in range(df.shape[0]):
    if(df.age[i]>33 and df.diabetes[i]==1):
        di['has diabetes']+=1
    elif(df.age[i]>33 and df.diabetes[i]==0):
        di['no diabetes']+=1
#print(di)
labels = di.keys()
sizes = di.values()

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%')
ax1.axis('equal')
plt.title('diabetic specimen')

plt.show()
```

```python
acc=accuracy_score(y_test,y_pred)
print("Accuracy is:",acc*100)
pre=precision_score(y_test,y_pred)
print("Precision:",pre*100)
rec=recall_score(y_test,y_pred)
print("Recall:",rec*100)
F1=f1_score(y_test,y_pred)
print("F1 Score:",F1*100)
```

```python
from sklearn.metrics import confusion_matrix, roc_curve, precision_recall_curve, roc_auc_score, average_precision_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from joblib import dump
x=df.iloc[:,0:16]
y=df["TenYearCHD"]
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=0)
model=LogisticRegression()
dump(model, 'logistic_regression_model.joblib')
model.fit(x_train,y_train)
y_pred=model.predict(x_test)


#cm = confusion_matrix(y_test, y_pred)
#print("Confusion Matrix:")
#print(cm)
fpr, tpr, _ = roc_curve(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred)
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc="lower right")
plt.show()

```
